```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

songs = pd.read_csv('C:/Users/user/Desktop/FYP/harmonicai/backend/database/spotify_songs.csv')
songs.head()
```

```python
# Define the linear transformation function
def convert_popularity_to_listens(popularity, old_min=0, old_max=100, new_min=0, new_max=32800):
    return ((popularity - old_min) / (old_max - old_min)) * (new_max - new_min) + new_min

# Apply the transformation to the 'track_popularity' column
songs['listen_count'] = songs['track_popularity'].apply(convert_popularity_to_listens).astype(int)

# Display the first few rows to verify the transformation
songs.head()
```

```python
# Drop unneccessary columns
songs = songs.drop(columns=['track_popularity', 'track_album_id', 'track_album_name', 'track_album_release_date', 'playlist_name', 'playlist_id', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 
            'instrumentalness', 'liveness', 'valence', 'tempo'])
songs.head()
```

```python
# Function to convert milliseconds to MM:SS format
def convert_ms_to_mmss(duration_ms):
    seconds = duration_ms // 1000
    minutes = seconds // 60
    seconds = seconds % 60
    return f"{minutes:02}:{seconds:02}"

# Apply the function to convert duration_ms to MM:SS format
songs['duration'] = songs['duration_ms'].apply(convert_ms_to_mmss)
songs = songs.drop(columns=['duration_ms'])

songs.head()
```

```python
songs['genre'] = songs['playlist_genre'] + ' ' + songs['playlist_subgenre']
songs = songs.drop(columns=['playlist_genre', 'playlist_subgenre'])
songs.head()
```

```python
# Store the new DataFrame in cleaned_spotify_songs.csv
songs.to_csv('C:/Users/user/Desktop/FYP/harmonicai/backend/database/cleaned_spotify_songs.csv', index=False)
```

```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import pandas as pd
import numpy as np
import re
import gensim.downloader as api
from gensim.parsing.preprocessing import remove_stopwords

songs = pd.read_csv('C:/Users/user/Desktop/FYP/harmonicai/backend/database/cleaned_spotify_songs.csv')
songs['track_name'] = songs['track_name'].astype(str)
song_names = songs['track_name']

# Load model
model = api.load('word2vec-google-news-300')

# Preprocess the sentences
song_names = [remove_stopwords(song.lower()).split() for song in song_names]


def sentence_vector(sentence, model):
    words = [word for word in sentence if word in model]
    if not words:
        return np.zeros(model.vector_size)
    return np.mean([model[word] for word in words], axis=0)


song_vectors = np.array([sentence_vector(song, model) for song in song_names])

# Save vector to a file
np.save('C:/Users/user/Desktop/FYP/harmonicai/backend/database/song_vector.npy', song_vectors)
```
